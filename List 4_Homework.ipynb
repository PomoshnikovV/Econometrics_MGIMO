```python
import numpy as np
import pandas as pd
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
```

# Задача 1.4

```python
df  =  pd . read_csv ( 'https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/wage2.csv' ) 
df 
```

```python
# подгонка прямой 
fitted_line = smf.ols(formula = 'np.log(wage)~age+IQ+married+south+urban', data = df).fit() 
fitted_line.params
fitted_line.summary(alpha=0.01).tables[1]
```

```python
alpha = 0.01
t_cr=stats.t.ppf(1-alpha/2, 935-1-5)
t_cr
```

```python
date_string = ''' Name  coef	std.err	t	P>|t|	[0.005 0.995]
Intercept	4.9740	0.165	30.064	0.000	4.547	5.401
age	0.0213	0.004	5.303	0.000	0.011	0.032
IQ	0.0082	0.001	9.751	0.000	0.006	0.010
married	0.2010	0.040	4.993	0.000	0.097	0.305
south	-0.0990	0.027	-3.690	0.000	-0.168	-0.030
urban	0.1750	0.028	6.331	0.000	0.104	0.246'''
```

```python
df_string = pd.read_csv(io.StringIO(date_string), sep='\s+')
df_string
```

```python
df_string['result'] = df_string.apply(lambda x: 'значимо' if abs(x['t'])>t_cr else 'не значимо',axis=1)
df_string
```

# Задача 2.1

```python
df  =  pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv' ) 
df 
```

```python
# подгонка прямой 
fitted_line = smf.ols(formula = 'np.log(output)~np.log(capital)+np.log(labour)', data = df).fit() 
fitted_line.summary(alpha=0.01).tables[1]
```

```python
t_nabl_capital = (0.2076-0.5)/0.017
t_cr = stats.t.ppf(1-0.01/2,569-2-1)
print('Коэффициент b_capital = 0.5') if abs(t_nabl_capital)<t_cr else print('Коэффициент b_capital значительно отличается от значения 0.5')
```
